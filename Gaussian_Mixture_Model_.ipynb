```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas import DataFrame
from sklearn import datasets
from sklearn.mixture import GaussianMixture
```

```python
iris = datasets.load_iris()
```

```python
features = iris.data[:, :2]

df = pd.DataFrame(features)
plt.scatter(df[0], df[1])

gmm = GaussianMixture(n_components = 3)
gmm.fit(df)

labels = gmm.predict(df)
df['labels']= labels

d0 = df[df['labels']== 0]
d1 = df[df['labels']== 1]
d2 = df[df['labels']== 2]
plt.show()

plt.scatter(d0[0], d0[1], c ='r')
plt.scatter(d1[0], d1[1], c ='g')
plt.scatter(d2[0], d2[1], c ='b')
plt.show()
```

```python
# printing the converged log-likelihood value
print(f'The mean: \n{gmm.means_}\n')
print(f'The covariance matrix: \n{gmm.covariances_}\n')
print(f'The converged log-likelihood value: {gmm.lower_bound_}\n')
print(f'The number of iterations needed to converge: {gmm.n_iter_}')
```
